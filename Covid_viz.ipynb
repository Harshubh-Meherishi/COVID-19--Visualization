```
import chardet
import pandas as pd

with open('/data.csv', 'rb') as f:
    result = chardet.detect(f.read(1024**2))  # or readline if the file is large


df =pd.read_csv('/data.csv', encoding=result['encoding'])
df.head(10)
```

```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
sns.set_style("darkgrid")
sns.FacetGrid(df,hue="geoId",height = 10).map(plt.scatter,"cases","deaths").add_legend()
   
plt.show()
```
